```python
dfmale = pd.read_csv('Indian-Male-Names.csv')
dffemale = pd.read_csv('Indian-Female-Names.csv')
```

```python
dffemale.head()
```

```python
gender = {'m':1,'f':0}
dfmale.gender = [gender[item] for item in dfmale.gender]
dffemale.gender = [gender[item] for item in dffemale.gender]
```

```python
dfmale.head()
```

```python
dfmale = dfmale.drop(columns=['race'])
dfmale.head()
```

```python
dffemale = dffemale.drop(columns=['race'])
```

```python
dffemale.head()
```

```python
df = pd.concat([dfmale,dffemale])
```

```python
df
```

```python
dffinal = df.sample(frac = 1)
dffinal.head()
```

```python
print(df.size)
print(dffinal.size)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import DictVectorizer
```

```python
names = dffinal['name']
names
```

```python
names[0]
```

```python
cv = CountVectorizer()
X = cv.fit_transform(names.values.astype('U'))
```

```python
cv.get_feature_names_out().tolist()
```

```python
Y = dffinal.gender
```

```python
x_train,x_test,y_train,y_test = train_test_split( X, Y,test_size = 0.33, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
sample = ['avinash']
vector = cv.transform(sample).toarray()
a = model.predict(vector).tolist()
if a == [1]:
  print('Male')
else :
  print('Female')
```
